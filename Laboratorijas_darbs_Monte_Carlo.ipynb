# Laboratorijas darbs - Monte Carlo

Dotā funkcija ir sinh(x/2).

No numpy tiek importēta funkcija ranodm

```python
from numpy import random
```

Tiek izvēlēts skaits punktiem, kurus rādis grafikā.

```python
N = 5000
```

Noteikts intervāls.

```python
a = 0
b = 5
```

Nejauši izvēlas x koordinātas.

```python
x = random.uniform(a,b,N)
```

Tiek parādīts cik nejauši ir izvēlti punkti. Ja izvēlēta x vērtība ir mazāka par 1, tā tiek ielikta pirmajā intervālā. Ja tā ir lielāka par 1, bet mazāka par divi, to liek otrajā intervālā. utt.

```python
k = [0, 0, 0, 0, 0]
for i in range(N):
    if x[i] < 1:
        k[0] = k[0] + 1
    elif x[i] < 2:
        k[1] = k[1] + 1
    elif x[i] < 3:
        k[2] = k[2] + 1
    elif x[i] < 4:
        k[3] = k[3] + 1
    else:
        k[4] = k[4] + 1
print(k)
```

No math tiek importēts sinh

```python
from math import sinh
```

Iegūst nejauši izvēlētas x koordinātas.

```python
y = random.uniform(a,b,N)
```

No matplotlib importē pyplot kā plt.

```python
from matplotlib import pyplot as plt
```

Visus nejauši izvēlētos punktus, kas ir mazāki par funkcijas vērtību pie attiecīgā sinh(x/2), iekrāso zaļā krāsā. Visus pārējos iekrāso sarkanā krāsā.

```python
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funkcija un taas integraalis (laukums starp funkciju un x ass)')
N1 = 0
for i in range(N):
    if y[i] < sinh(x[i]/2):
        plt.plot(x[i],y[i],'go')
        N1 = N1 + 1
    else:
        plt.plot(x[i],y[i],'ro')
plt.show()
```

Iegūst kopējo taisstūra laukumu.

To reizinot ar atteicību starp zaļajiem punktiem un visiem punktiem kopā, iegūst laukumu zem funkcijas, jeb integrāļa vērtību.

```python
S_zinaamais = (b-a) * (b-a)
S_nezinaamais = 1. * S_zinaamais * N1 / N
print(S_nezinaamais)
```
